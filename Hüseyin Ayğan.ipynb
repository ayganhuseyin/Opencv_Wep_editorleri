```python
5-# Kenarları algılama

# OpenCV kütüphanesini çağırma
import cv2

# Görüntüyü okuma
image = cv2.imread('resim.jpg')

# Görüntüyü gri tonlamaya dönüştürme
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Kenarları algılama
edges = cv2.Canny(gray_image, threshold1=30, threshold2=100)

# Kenarları ekranda gösterme
cv2.imshow('Kenarlar', edges)

# Bir tuşa basılana kadar bekletme
cv2.waitKey(0)

# Tüm pencereleri kapatma
cv2.destroyAllWindows()
```

```python

6-# Kenarları ekranda gösterme

# OpenCV kütüphanesini çağırma
import cv2

# Görüntüyü okuma
image = cv2.imread('resim.jpg')

# Görüntüyü gri tonlamaya dönüştürme
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Kenarları algılama
edges = cv2.Canny(gray_image, threshold1=30, threshold2=100)

# Kenarları orijinal görüntü üzerinde gösterme
edge_image = cv2.cvtColor(edges, cv2.COLOR_GRAY2BGR)  # Kenarları renkli görüntüye dönüştürme
result = cv2.addWeighted(image, 0.7, edge_image, 0.3, 0)  # Kenarları orijinal görüntüyle birleştirme

# Görüntüyü ekranda gösterme
cv2.imshow('Kenarlar', result)

# Bir tuşa basılana kadar bekletme
cv2.waitKey(0)

# Tüm pencereleri kapatma
cv2.destroyAllWindows()
```

```python
7-# Yüz tanıma

# OpenCV kütüphanesini çağırma
import cv2

# CascadeClassifier ile yüz tanıma sınıflandırıcısını yükleme
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

# Görüntüyü okuma
image = cv2.imread('resim.jpg')

# Görüntüyü gri tonlamaya dönüştürme
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Yüzleri tanıma
faces = face_cascade.detectMultiScale(gray_image, scaleFactor=1.1, minNeighbors=5, minSize=(30, 30))

# Her bir yüzü çerçeve içine alma
for (x, y, w, h) in faces:
    cv2.rectangle(image, (x, y), (x+w, y+h), (255, 0, 0), 2)

# Görüntüyü ekranda gösterme
cv2.imshow('Yüz Tanıma', image)

# Bir tuşa basılana kadar bekletme
cv2.waitKey(0)

# Tüm pencereleri kapatma
cv2.destroyAllWindows()
```

```python

8-# Yüzleri ekranda gösterme

# OpenCV kütüphanesini çağırma
import cv2

# CascadeClassifier ile yüz tanıma sınıflandırıcısını yükleme
face_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')

# Görüntüyü okuma
image = cv2.imread('resim.jpg')

# Görüntüyü gri tonlamaya dönüştürme
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Yüzleri tanıma
faces = face_cascade.detectMultiScale(gray_image, scaleFactor=1.1, minNeighbors=5, minSize=(30, 30))

# Her bir yüzü çerçeve içine alma ve görselleştirme
for (x, y, w, h) in faces:
    cv2.rectangle(image, (x, y), (x+w, y+h), (255, 0, 0), 2)

# Görüntüyü ekranda gösterme
cv2.imshow('Yüzler', image)

# Bir tuşa basılana kadar bekletme
cv2.waitKey(0)

# Tüm pencereleri kapatma
cv2.destroyAllWindows()
```

```python

9-#Görüntü Bulanıklaştırma

# OpenCV kütüphanesini çağırma
import cv2

# Görüntüyü okuma
image = cv2.imread('resim.jpg')

# Görüntüyü bulanıklaştırma
blurred_image = cv2.GaussianBlur(image, (15, 15), 0)

# Bulanıklaştırılmış görüntüyü ekranda gösterme
cv2.imshow('Bulanıklaştırılmış Görüntü', blurred_image)

# Bir tuşa basılana kadar bekletme
cv2.waitKey(0)

# Tüm pencereleri kapatma
cv2.destroyAllWindows()
```
